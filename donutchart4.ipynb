```python
import colorsys
import matplotlib.pyplot as plt
import numpy as np

def hex_to_rgb(hexc):
    '''Convert hex triple color to RGB tuple.'''
    hexc = hexc.lstrip('#')
    return tuple(int(hexc[idx:idx+2], 16) for idx in (0, 2 ,4))

def normalize (rgb):
    '''Normalize RGB color to be in range [0, 1].'''
    return rgb[0] / 255.0, rgb[1] / 255.0, rgb[2] / 255.0

def saturate(hexc, factor=0.6667):
    '''Change saturation of an RGB color using conversion to HSV.'''
    (hue, sat, val) = colorsys.rgb_to_hsv(*normalize (hex_to_rgb(hexc)))
    sat = sat * factor
    return colorsys.hsv_to_rgb (*(hue, sat, val))

def formatter(pct):
    return '{:0.1f}%'.format(pct)

fig, ax = plt.subplots()
ax.axis('equal')

colors=['#66FF66', '#FF9999']

outside, _, _ = ax.pie([196, 24], radius=1, autopct=formatter, pctdistance=0.8,
                       startangle=90, colors=colors)

inside, _ = ax.pie([45, 87], radius=0.67**1,
                   startangle=90, colors=[saturate(color, 0.5**1) for color in colors])
             
inside2, _ = ax.pie([12, 33], radius=0.67**2,
                    startangle=90, colors=[saturate(color, 0.5**2) for color in colors])
             
inside3, _ = ax.pie([33, 9], radius=0.67**3,
                    startangle=90, colors=[saturate(color, 0.5**3) for color in colors])
             
inside4, _ = ax.pie([10, 24], radius=0.67**4,
                    startangle=90, colors=[saturate(color, 0.5**4) for color in colors])
             
#plt.setp(inside + inside2 + outside, width=width, edgecolor='white')

ax.annotate('2017-11-19',
            xy=(0.8, 0), xycoords='data',
            xytext=(290, 110), textcoords='figure pixels', ha='left', alpha=0.67**1,
            arrowprops=dict(arrowstyle="->", connectionstyle='angle3,angleA=0,angleB=-45', alpha=0.67**1))

ax.annotate('2017-11-18',
            xy=(0.55, 0), xycoords='data',
            xytext=(290, 90), textcoords='figure pixels', ha='left', alpha=0.67**2,
            arrowprops=dict(arrowstyle="->", connectionstyle='angle3,angleA=0,angleB=-45', alpha=0.67**2))

ax.annotate('2017-11-17',
            xy=(0.35, 0), xycoords='data',
            xytext=(290, 70), textcoords='figure pixels', ha='left', alpha=0.67**3,
            arrowprops=dict(arrowstyle="->", connectionstyle='angle3,angleA=0,angleB=-45', alpha=0.67**3))

ax.annotate('2017-11-16',
            xy=(0.20, 0), xycoords='data',
            xytext=(290, 50), textcoords='figure pixels', ha='left', alpha=0.67**4,
            arrowprops=dict(arrowstyle="->", connectionstyle='angle3,angleA=0,angleB=-45', alpha=0.67**4))

ax.annotate('2017-11-15',
            xy=(0.1, 0), xycoords='data',
            xytext=(290, 30), textcoords='figure pixels', ha='left', alpha=0.67**5,
            arrowprops=dict(arrowstyle="->", connectionstyle='angle3,angleA=0,angleB=-45', alpha=0.67**5))

plt.show()
```
